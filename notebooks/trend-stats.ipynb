```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def plot_leng_ret(key, statsT, statsF):
    props = dict(boxstyle='round', facecolor='white', alpha=0.5) # text boxes
    
    fig, axs = plt.subplots(ncols=4, figsize=(16,3))
    axs[0].scatter(statsT.leng, statsT['max'], color = 'green'); axs[0].set_title(f'{key} length - max %'); axs[0].set_xlabel(f'{tf} candles')
    axs[0].set_yscale('log')
    
    axs[0].text(0.55, 0.15, f'# datapoints: {len(statsT)}', transform=axs[0].transAxes, fontsize=10,
        verticalalignment='top', bbox=props)
    
    axs[1].scatter(statsT.leng, statsT['pct'], color = 'green'); axs[1].set_title(f'{key} length - pct %'); axs[1].set_xlabel(f'{tf} candles')
    axs[1].set_yscale('log')

    axs[2].scatter(statsF.leng, statsF['max'], color = 'red'); axs[2].set_title(f'{key} length - max %'); axs[0].set_xlabel(f'{tf} candles')
    axs[2].set_yscale('log')
    
    axs[2].text(0.55, 0.95, f'# datapoints: {len(statsF)}', transform=axs[2].transAxes, fontsize=10,
        verticalalignment='top', bbox=props)
    
    axs[3].scatter(statsF.leng, statsF['pct'], color = 'red'); axs[3].set_title(f'{key} length - pct %'); axs[1].set_xlabel(f'{tf} candles')
    axs[3].set_yscale('log')

    plt.tight_layout()
    plt.show()
```

```python
tf = '1d'; key = 'doge'; pair = 'usd'
slow_ma = 200; fast_ma = 50
in_filename = f"../data/{tf}/{pair}/{key}/trends_{tf}_{fast_ma}-{slow_ma}.csv"
stats = pd.read_csv(in_filename)
```

```python
stats
```

```python
statsT = stats[ stats.GC == True ]
statsF = stats[ stats.GC == False ]
plot_leng_ret(key, statsT, statsF)
```
